# Data processing walkthrough

This notebook follows the same data path as the project code:

1. Read one raw CSV file with pandas.
2. Normalize and resample the time column to 100 Hz.
3. Extract the model features used by `scripts/hdf5.py`.
4. Convert the continuous trajectory into sliding-window samples `X` and `Y`.
5. Inspect the saved HDF5 files.
6. Load the HDF5 file through the project's PyTorch `Dataset` and `DataLoader`.

The cells are intentionally small so you can run them one by one.

## 0. Imports and repository paths

Run this cell first. It finds the repository root even if Jupyter starts inside the `notebooks/` folder.

````python
from pathlib import Path
import sys

import h5py
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


def find_repo_root(start: Path) -> Path:
    start = start.resolve()
    for candidate in [start, *start.parents]:
        if (candidate / "scripts" / "hdf5.py").exists() and (candidate / "resources").exists():
            return candidate
    raise RuntimeError("Could not find repository root. Start Jupyter inside this repository.")


REPO_ROOT = find_repo_root(Path.cwd())
SCRIPTS_DIR = REPO_ROOT / "scripts"
if str(SCRIPTS_DIR) not in sys.path:
    sys.path.insert(0, str(SCRIPTS_DIR))

print("Repository root:", REPO_ROOT)
````

## 1. Choose the dataset and split

`DATASET` is set to `neurobem` here so the notebook follows the NeuroBEM data path first. You can later switch it to `pi_tcn` if needed.

````python
DATASET = "neurobem"  # Switch to "pi_tcn" later if needed.
SPLIT = "train"     # Usually "train", "valid", or "test".

history_length = 20
unroll_length = 10
DATA_DTYPE = np.float32

DATA_DIR = REPO_ROOT / "resources" / "data" / DATASET
SPLIT_DIR = DATA_DIR / SPLIT

print("Data directory:", DATA_DIR)
print("Split directory:", SPLIT_DIR)
print("Split exists:", SPLIT_DIR.exists())
````

## 2. List raw CSV files

The preprocessing script loops over CSV files in folders such as `resources/data/pi_tcn/train/`.

````python
csv_files = sorted(SPLIT_DIR.glob("*.csv"))
print("Number of CSV files:", len(csv_files))
for path in csv_files[:10]:
    print(path.relative_to(REPO_ROOT))

if not csv_files:
    raise FileNotFoundError(f"No CSV files found in {SPLIT_DIR}")
````

## 3. Read one CSV with pandas

This is the same first step as `pd.read_csv(csv_file_path)` in `scripts/hdf5.py`.

````python
csv_path = csv_files[0]
raw = pd.read_csv(csv_path)

print("Using file:", csv_path.relative_to(REPO_ROOT))
print("Shape:", raw.shape)
display(raw.head())
````

````python
column_summary = pd.DataFrame({
    "column": raw.columns,
    "dtype": raw.dtypes.astype(str).values,
    "missing": raw.isna().sum().values,
})
display(column_summary)
````

## 4. Normalize time and resample to 100 Hz

This mirrors these lines in `scripts/hdf5.py`:

```python
data['t'] = data['t'] - data['t'].values[0]
data['t'] = pd.to_datetime(data['t'], unit='s')
data.set_index('t', inplace=True)
data = data.resample('0.01s').mean()
data.reset_index(inplace=True)
```

````python
data = raw.copy()

# Make the first timestamp equal to 0 seconds.
data["t"] = data["t"] - data["t"].values[0]

# Convert seconds to a datetime-like index so pandas can resample by time.
data["t"] = pd.to_datetime(data["t"], unit="s")
data = data.set_index("t")

# 0.01 seconds = 100 Hz.
resampled = data.resample("0.01s").mean().reset_index()

print("Raw shape:", raw.shape)
print("Resampled shape:", resampled.shape)
display(resampled.head())
````

````python
plot_column = "v_x" if DATASET == "pi_tcn" else "vel x"

raw_time_seconds = raw["t"] - raw["t"].iloc[0]
resampled_time_seconds = (resampled["t"] - resampled["t"].iloc[0]).dt.total_seconds()

fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(raw_time_seconds, raw[plot_column], ".", alpha=0.25, label="raw")
ax.plot(resampled_time_seconds, resampled[plot_column], "-", linewidth=1.2, label="resampled 100 Hz")
ax.set_title(f"Before and after resampling: {plot_column}")
ax.set_xlabel("time [s]")
ax.set_ylabel(plot_column)
ax.legend()
ax.grid(True, alpha=0.3)
````

## 5. Extract the features used by the model

This cell uses the same feature-selection logic as `extract_data(...)` in `scripts/hdf5.py`. It chooses different CSV columns for `pi_tcn` and `neurobem`.

````python
FEATURE_NAMES_BY_DATASET = {
    "pi_tcn": [
        "v_x", "v_y", "v_z",
        "q_w", "q_x", "q_y", "q_z",
        "w_x", "w_y", "w_z",
        "u_0", "u_1", "u_2", "u_3",
    ],
    "neurobem": [
        "vel x", "vel y", "vel z",
        "quat w", "quat x", "quat y", "quat z",
        "ang vel x", "ang vel y", "ang vel z",
        "mot 1", "mot 2", "mot 3", "mot 4",
    ],
}

def extract_data_like_project(data, dataset_name):
    if dataset_name == "pi_tcn":
        velocity_data = data[["v_x", "v_y", "v_z"]].values
        attitude_data = data[["q_w", "q_x", "q_y", "q_z"]].values
        angular_velocity_data = data[["w_x", "w_y", "w_z"]].values
        control_data = data[["u_0", "u_1", "u_2", "u_3"]].values * 0.001
    elif dataset_name == "neurobem":
        velocity_data = data[["vel x", "vel y", "vel z"]].values
        attitude_data = data[["quat w", "quat x", "quat y", "quat z"]].values
        angular_velocity_data = data[["ang vel x", "ang vel y", "ang vel z"]].values
        control_data = data[["mot 1", "mot 2", "mot 3", "mot 4"]].values * 0.001
    else:
        raise ValueError(f"Unknown dataset: {dataset_name}")
    return velocity_data, attitude_data, angular_velocity_data, control_data


feature_names = FEATURE_NAMES_BY_DATASET[DATASET]

velocity_data, attitude_data, angular_velocity_data, control_data = extract_data_like_project(resampled, DATASET)
data_np = np.hstack((velocity_data, attitude_data, angular_velocity_data, control_data)).astype(DATA_DTYPE, copy=False)

print("velocity_data:", velocity_data.shape)
print("attitude_data:", attitude_data.shape)
print("angular_velocity_data:", angular_velocity_data.shape)
print("control_data:", control_data.shape)
print("combined data_np:", data_np.shape, data_np.dtype)

display(pd.DataFrame(data_np[:5], columns=feature_names))
````

## 6. Build sliding-window samples

For every possible start index `i`, the code creates:

- `X[i]`: the past `history_length` states.
- `Y[i]`: the next `unroll_length` states to predict.

So one continuous trajectory becomes many supervised learning examples.

````python
num_samples = data_np.shape[0] - history_length - unroll_length
if num_samples <= 0:
    raise ValueError("This CSV is too short for the chosen history_length and unroll_length.")

X = np.zeros((num_samples, history_length, data_np.shape[1]), dtype=DATA_DTYPE)
Y = np.zeros((num_samples, unroll_length, data_np.shape[1]), dtype=DATA_DTYPE)

for i in range(num_samples):
    X[i, :, :] = data_np[i : i + history_length, :]
    Y[i, :, :] = data_np[i + history_length : i + history_length + unroll_length, :]

print("X shape:", X.shape, "= [num_samples, history_length, features]")
print("Y shape:", Y.shape, "= [num_samples, unroll_length, features]")
````

````python
sample_idx = 0

print("One training example:")
print("X[sample_idx] shape:", X[sample_idx].shape)
print("Y[sample_idx] shape:", Y[sample_idx].shape)

print("History window X:")
display(pd.DataFrame(X[sample_idx], columns=feature_names).head())

print("Prediction target Y:")
display(pd.DataFrame(Y[sample_idx], columns=feature_names).head())
````

````python
sample_idx = 0
feature = 0

fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(data_np[: history_length + unroll_length + 20, feature], label=feature_names[feature])
ax.axvspan(0, history_length - 1, color="tab:blue", alpha=0.12, label="X history")
ax.axvspan(history_length, history_length + unroll_length - 1, color="tab:orange", alpha=0.18, label="Y target")
ax.set_title("How one sliding-window sample is cut from a trajectory")
ax.set_xlabel("time step after resampling")
ax.grid(True, alpha=0.3)
ax.legend()
````

## 7. Inspect saved HDF5 files

`scripts/hdf5.py` saves arrays under two keys:

- `inputs`: all `X` samples.
- `outputs`: all `Y` samples.

For `train` and `valid`, each split is usually stored as one file: `train.h5` or `valid.h5`.

````python
h5_files = sorted(SPLIT_DIR.glob("*.h5"))
print("HDF5 files in split:", len(h5_files))
for path in h5_files[:20]:
    print(path.relative_to(REPO_ROOT))
````

````python
if h5_files:
    h5_path = h5_files[0]
else:
    h5_path = None
    print("No HDF5 file found for this split. Run scripts/hdf5.py first if you want to inspect saved data.")

if h5_path is not None:
    with h5py.File(h5_path, "r") as hf:
        print("Using:", h5_path.relative_to(REPO_ROOT))
        print("Keys:", list(hf.keys()))
        print("inputs shape:", hf["inputs"].shape, hf["inputs"].dtype)
        print("outputs shape:", hf["outputs"].shape, hf["outputs"].dtype)
        print("input dimension labels:", [hf["inputs"].dims[i].label for i in range(hf["inputs"].ndim)])
        print("output dimension labels:", [hf["outputs"].dims[i].label for i in range(hf["outputs"].ndim)])
````

## 8. Load through the project's PyTorch Dataset/DataLoader

This mirrors what `scripts/train.py` does through `load_dataset(...)`.

````python
if h5_path is None:
    print("Skip DataLoader demo because no HDF5 file is available for this split.")
else:
    try:
        from types import SimpleNamespace
        from dynamics_learning.data import load_dataset
    except ModuleNotFoundError as exc:
        print("Skip DataLoader demo because a dependency is missing:", exc)
    else:
        args = SimpleNamespace(
            history_length=history_length,
            unroll_length=unroll_length,
            batch_size=4,
            shuffle=False,
            model_type="tcn",
        )

        dataset_obj, loader = load_dataset(
            mode="notebook demo",
            data_path=str(h5_path.parent),
            hdf5_file=h5_path.name,
            args=args,
            num_workers=0,
            pin_memory=False,
        )
        batch_x, batch_y = next(iter(loader))
        print("Dataset length:", len(dataset_obj))
        print("Batch X:", batch_x.shape, batch_x.dtype)
        print("Batch Y:", batch_y.shape, batch_y.dtype)
````
